```python
drop_columns=['Start Date-Subquestion',
       'End Date-Subquestion', 'Email Address-Subquestion',
       'First Name-Subquestion', 'Last Name-Subquestion',
       'Custom Data 1-Subquestion',]
dataset_modified=dataset_modified.drop(columns=drop_columns)
```

```python
list(dataset_modified.columns)[8:]
```

```python
id_vars=list(dataset_modified.columns)[:8]
value_vars=list(dataset_modified.columns)[8:]
```

```python
dataset_melted=dataset_modified.melt(id_vars=id_vars, value_vars=value_vars, var_name="Question + Subquestion", value_name="Answer")
```

```python
question_input=pd.read_excel(r'C:\Users\daniel\OneDrive - University of Birmingham (1)\Python Scripts\Data - Survey Monkey Output.xlsx', sheet_name="Question")
```

```python
question=question_input.copy()
question.drop(columns=["Raw Question","Raw Subquestion","Subquestion"],inplace=True)
```

```python
dataset_merged=pd.merge(left=dataset_melted, right=question, how="left", left_on="Question + Subquestion", right_on="Question + Subquestion")
```

```python
respondents=dataset_merged[dataset_merged["Answer"].notna()].groupby("Question")["Respondent ID-Subquestion"].nunique().reset_index()
respondents.rename(columns={"Respondent ID-Subquestion":"Respondents"},inplace=True)
```

```python
dataset_merged_two=pd.merge(left=dataset_merged, right=respondents, how="left", left_on="Question", right_on="Question")
```

```python
respondentstest=dataset_merged[dataset_merged["Answer"].notna()].groupby("Question")#["Respondent ID-Subquestion"].nunique().reset_index()
#respondents.rename(columns={"Respondent ID-Subquestion":"Respondents"},inplace=True)
```

```python
dataset_merged[dataset_merged["Answer"].notna()]
```

```python
same_answer=dataset_merged_two
same_answer=same_answer.groupby(["Question + Subquestion","Answer"])["Respondent ID-Subquestion"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID-Subquestion":"Same answer"},inplace=True)
```

```python
dataset_merged_three=pd.merge(left=dataset_merged_two, right=same_answer, how="left", left_on=["Question + Subquestion","Answer"], right_on=["Question + Subquestion","Answer"])
dataset_merged_three["Same answer"].fillna(0,inplace=True)
```
